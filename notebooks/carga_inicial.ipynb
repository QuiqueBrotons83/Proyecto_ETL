```python
# Importamos las librerias necesarias para empezar
import pandas as pd
import numpy as np
import psycopg2 as ps
import pyarrow 
```

```python
# Creamos un df sobre el archivo parquet y guardamos el dataset crudo.

df_raw = pd.read_parquet("../data/reservas_hoteles.parquet",engine='pyarrow')
df_raw
```

```python
# Guardamos el dataset como datos crudo y así emprezamos a trabajar con el, con un eda
df_raw.to_parquet("../data/raw_hoteles.parquet")
```

```python
# Empezamos con el EDA
df_hoteles = df_raw
df_hoteles
```

```python

df_hoteles.duplicated().sum()
df_hoteles = df_hoteles.drop_duplicates()
```

```python
# comprobamos que los id_hotel no estan bien. Vemos la columna hotel para ver cuantos valores unicos tienes.
df_hoteles['nombre_hotel'].unique().T
```

### Vemos que tenemos 19 hoteles en total

```python
# Creamos 2 dataframe que separen los hoteles de la competencia con los propios
df_propios = df_hoteles[df_hoteles['competencia']==False]
df_propios
```

```python
# Eliminar la columna 'id_hotel' actual
df_propios = df_propios.drop(columns=['id_hotel'])

# Crear un diccionario que asocie cada hotel único con un nuevo ID
unique_hotels = df_propios['nombre_hotel'].unique()
hotel_id_map = {hotel: idx + 1 for idx, hotel in enumerate(unique_hotels)}
print(hotel_id_map)

# Asignar los nuevos IDs a la columna 'id_hotel'
df_propios['id_hotel'] = df_propios['nombre_hotel'].map(hotel_id_map)

```

```python
df_hoteles_competencia = df_hoteles[df_hoteles['competencia']==True]
df_hoteles_competencia
```

```python
# Eliminar la columna 'id_cliente'
if 'id_cliente' in df_hoteles.columns:
    df_hoteles = df_hoteles.drop(columns=['id_cliente'])

# Crear una nueva columna 'id_cliente' basada en el correo electrónico
df_hoteles['id_cliente'] = df_hoteles['mail'].factorize()[0] + 1000

df_hoteles['id_cliente'].unique()
```

```python
# Miro que tal han quedado los datos despues de elminar el id_cliente y crear uno nuevo
valoracion = df_propios.groupby('nombre_hotel')["estrellas"].mean()
valoracion
```

```python
precio = df_propios.groupby('nombre_hotel')["precio_noche"].mean()
precio
```

```python
propios_merge = pd.merge(df_propios, valoracion, on='nombre_hotel')
propios_merge
```

```python
propios_merge = pd.merge(propios_merge, precio, on='nombre_hotel')
propios_merge.info()
```

```python
propios_merge = propios_merge.drop(columns=['estrellas_x', 'precio_noche_x'])
```

```python
propios_merge = propios_merge.rename(columns={'estrellas_y': 'estrellas', 'precio_noche_y': 'precio_medio'})
propios_merge.info()
```

```python
scrapeo = pd.read_csv("../data/scrapeo_hoteles.csv", parse_dates= ['fecha_reserva'])
scrapeo.info()
```

```python
df_hoteles_competencia["id_hotel"].unique()
```

```python
scrapeo["id_hotel"] = df_hoteles_competencia["id_hotel"].unique()
scrapeo
```

```python
merge_competencia = pd.merge(df_hoteles_competencia, scrapeo, on='id_hotel')
merge_competencia.info()
```

```python
# Elimino las columnas que no me interesan
merge_competencia = merge_competencia.drop(columns= ["fecha_reserva_x", "nombre_hotel_x", "estrellas_x", "precio_noche"])
merge_competencia.info()
```

```python
merge_competencia = merge_competencia.rename(columns={"fecha_reserva_y": "fecha_reserva", "nombre_hotel_y": "nombre_hotel", "estrellas_y": "estrellas", "precio": "precio_medio"})
```

```python
merge_competencia = merge_competencia.reindex(columns=propios_merge.columns)
```

```python
hoteles_unidos = pd.concat([propios_merge, merge_competencia],axis=0)
hoteles_unidos.info()
```

```python
# Veo que hay columnas de tipo fecha y decido convertirlas a tipo fecha
hoteles_unidos['fecha_reserva'] = pd.to_datetime(hoteles_unidos['fecha_reserva'], errors='coerce')
hoteles_unidos['inicio_estancia'] = pd.to_datetime(hoteles_unidos['inicio_estancia'], errors='coerce')
hoteles_unidos['final_estancia'] = pd.to_datetime(hoteles_unidos['final_estancia'], errors='coerce')
```
